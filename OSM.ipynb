```python
import osmnx as ox
import geopandas as gpd

# Set the query tags and location
bus_stop_tags = {"highway": "bus_stop"}
place = "Greater London"

# Get the geographic boundary of Greater London
gdf_greater_london = ox.geocode_to_gdf(place)

# Extract the geometry of Greater London
greater_london_geometry = gdf_greater_london.geometry.iloc[0]

# Query bus stops within the geometry of Greater London
bus_stops = ox.geometries_from_polygon(greater_london_geometry, tags=bus_stop_tags)

# Convert the result to a GeoDataFrame
bus_stops_gdf = gpd.GeoDataFrame(bus_stops)
```

```python
bus_stops_gdf.to_file("london_bus_stops.geojson", driver="GeoJSON")
```

## shop

```python
shop_tags = {"shop": True}
shops = ox.geometries_from_polygon(greater_london_geometry, shop_tags)

shops_gdf = gpd.GeoDataFrame(shops)
```
